```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/startup_funding.csv')
```

```python
sorted(df['Startup Name'].unique().tolist())
```

```python
df['Investors Name']=df['Investors Name'].fillna('Undisclosed')
```

```python
sorted(df['Investors Name'].unique().tolist())
```

```python
df.drop(columns=['Remarks'],inplace=True)
```

```python
df.set_index('Sr No',inplace=True)
```

```python
df.rename(columns={
        'Date dd/mm/yyyy':'date',
        'Startup Name':'Startup',
        'Industry Vertical':'Vertical',
        'SubVertical':'subvertical',
        'City  Location':'city',
        'Investors Name':'investors',
        'InvestmentnType':'round',
        'Amount in USD':'amount'
},inplace=True)
```

```python
df['amount']=df['amount'].fillna('0')
```

```python
df['amount']=df['amount'].str.replace(',','')
```

```python
df['amount']=df['amount'].str.replace('undisclosed','0')
```

```python
df['amount']=df['amount'].str.replace('Undisclosed','0')
```

```python
df['amount']=df['amount'].str.replace('unknown','0')
```

```python
df=df[df['amount'].str.isdigit()]
```

```python
df['amount']=df['amount'].astype(float)
```

```python
def to_inr(dollor):
    inr=dollor*82.5
    return inr/10000000
```

```python
df['amount']=df['amount'].apply(to_inr)
```

```python
df['date']=pd.to_datetime(df['date'],errors='coerce',format='mixed')
```

```python
df.info()
```

```python
df=df.dropna(subset=['date','Startup','Vertical','city','investors','round'])
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv',index=False)
```
